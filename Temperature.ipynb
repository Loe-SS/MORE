```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import netCDF4
import xarray as xr
```

```python
ds = xr.open_dataset('data/bteca478.nc')
ds
```

```python
df = ds.to_dataframe().reset_index()
df.to_csv('data/temperature.csv', index=False)


df.plot(x='time', y='tg', color = 'darkorange', figsize=(15,5))
plt.ylabel('Temperature (°C)')
plt.grid()

display(df.head())
```

```python
zero_count = (df['rr'] == 0).sum()
NaN_count = df['rr'].isna().sum()
print(f"Number of zero values: {zero_count}")
print(f"Number of NaN values: {NaN_count}")
```
